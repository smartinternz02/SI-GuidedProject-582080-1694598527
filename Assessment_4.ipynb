```python
data['Attrition'] = data['Attrition'].replace({'No':0,'Yes':1})
```

```python
data.head()
```

```python
data['OverTime'] = data['OverTime'].map({'No':0,'Yes':1})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders = {}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

```python
data.head()
```

Splitting of data into Dependent and Independent Variables

```python
X = data.drop(['Attrition','Over18'], axis=1)
y = data['Attrition']
```

```python
X.head()
```

```python
y.head()
```

```python
y.shape
```

```python
X.shape
```

Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(X),columns=X.columns)
```

Splitting data into Training and Testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled,y, test_size=0.2, random_state=0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

Model Building Using
1.LogisticRegression
2.DecisionTreeClassifier
3.RandomForestClassifier

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr
```

```python
lr.fit(X_train,y_train)
```

```python
lrpred=lr.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
dtpred=dt.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
rfpred=rf.predict(X_test)
```

Checking of Performance metrics

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score,classification_report
```

LOGISTIC REGRESSION

```python
accuracy_score(lrpred,y_test)
```

```python
confusion_matrix(lrpred,y_test)
```

```python
from pandas.core.reshape.pivot import crosstab
crosstab(lrpred,y_test)
```

```python
print(classification_report(lrpred,y_test))
```

```python
lrprob=lr.predict_proba(X_test)[:,1]
```

```python
fpr,tpr,threshsholds = roc_curve(y_test,lrprob)
```

```python
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```
